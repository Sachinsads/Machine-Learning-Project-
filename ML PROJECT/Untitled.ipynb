```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
print("Libraries imported successfully.")
try:
file_path = r"C:\Users\91887\Desktop\tourist_places.csv"
df = pd.read_csv(file_path)
df_original = df.copy()
print("Dataset loaded successfully!\n")
print("Original Data Sample:\n", df.head(), "\n")
except FileNotFoundError:
print(f"Error: The file was not found at {file_path}")
print("Please make sure 'tourist_places.csv' is on your Desktop.")
exit()
df.fillna(method='ffill', inplace=True)
le_country = LabelEncoder()
le_category = LabelEncoder()
df['Country_Code'] = le_country.fit_transform(df['Country'])
df['Category_Code'] = le_category.fit_transform(df['Category'])
scaler = MinMaxScaler()
df[['AverageRating', 'VisitorsPerYear']] = scaler.fit_transform(df[['AverageRating',
'VisitorsPerYear']])
print("Data preprocessing complete.\n")
print("Processed Data Sample:\n", df.head(), "\n")
features = df[['AverageRating', 'VisitorsPerYear', 'Country_Code', 'Category_Code']]
similarity_matrix = cosine_similarity(features)
print("Cosine similarity matrix calculated.\n")
def recommend_place(place_name, num_recommendations=5):
if place_name not in df['PlaceName'].values:
print(f"Error: '{place_name}' was not found in the dataset.")
return
idx = df[df['PlaceName'] == place_name].index[0]
similarity_scores = list(enumerate(similarity_matrix[idx]))
similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
top_similar_places_indices = [i[0] for i in similarity_scores[1:num_recommendations+1]]
print(f"--- Top {num_recommendations} Recommendations similar to '{place_name}' ---")
for i, place_idx in enumerate(top_similar_places_indices):
place_info = df.iloc[place_idx]
print(f"{i+1}. {place_info['PlaceName']} ({place_info['Country']})")
print("-" * 50 + "\n")
recommend_place("Goa Beaches")
print("Generating advanced visualizations... (This may open several windows)")
fig1 = px.scatter(
df,
x='AverageRating',
y='VisitorsPerYear',
size=df_original['VisitorsPerYear'],
color='Category',
hover_name='PlaceName',
text='PlaceName',
title='Quadrant Analysis: Rating vs. Popularity',
labels={'AverageRating': 'Average Rating (Normalized)', 'VisitorsPerYear': 'Visitors Per Year(Normalized)'},
template='plotly_white'
)
fig1.update_traces(textposition='top center', textfont_size=8)
fig1.update_layout(showlegend=True, title_x=0.5)
fig1.show()
plt.figure(figsize=(12, 7))
sns.violinplot(x='Category', y='AverageRating', data=df, inner='quartile', palette='pastel')
sns.swarmplot(x='Category', y='AverageRating', data=df, color='k', alpha=0.7)
plt.title('Rating Density and Distribution by Category', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Average Rating (Normalized)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
country_metrics = df_original.groupby('Country').agg(
AverageRating=('AverageRating', 'mean'),
NumPlaces=('PlaceName', 'count')
).reset_index()
fig3 = px.choropleth(
country_metrics,
locations='Country',
locationmode='country names',
color='AverageRating',
hover_name='Country',
hover_data={'NumPlaces': True, 'AverageRating': ':.2f'},
color_continuous_scale=px.colors.sequential.Plasma,
title='Global Heatmap of Average Tourist Ratings'
)
fig3.update_layout(title_x=0.5)
fig3.show()
fig4 = px.treemap(
df_original,
path=[px.Constant("All Destinations"), 'Country', 'Category'],
values='VisitorsPerYear',
color='AverageRating',
hover_name='PlaceName',
color_continuous_scale='YlGnBu',
title='Hierarchical View: Visitors & Ratings by Country and Category'
)
fig4.update_layout(margin=dict(t=50, l=25, r=25, b=25))
fig4.show()
top_places = df_original.sort_values(by='AverageRating', ascending=False).head(5)
print("\n--- Top 5 Highest Rated Tourist Destinations Overall ---")
print(top_places[['PlaceName', 'Country', 'Category', 'AverageRating']])
print("-" * 55)
output_filename = "top_recommended_places.csv"
top_places.to_csv(output_filename, index=False)
print(f"\nAnalysis complete. Top recommendations saved to '{output_filename}'")
```
